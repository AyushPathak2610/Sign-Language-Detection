```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import os
import time
import mediapipe as mp
```

```python
mp_holistic = mp.solutions.holistic # holistic model
mp_drawing = mp.solutions.drawing_utils # drawing utilities
```

```python
def mediapipe_detection(image, model):
    image = cv.cvtColor(image, cv.COLOR_BGR2RGB) # Color conversion BGR 2 BGR
    image.flags.writeable = False                # Image is no longer writeable
    results = model.process(image)               # Make prediction
    image.flags.writeable = True                 # Image is now writeable
    image = cv.cvtColor(image, cv.COLOR_RGB2BGR) # COLOR CONVERSION RGB 2 BGR
    return image, results
```

```python
def draw_landmarks(image,results):
    mp_drawing.draw_landmarks(image, results.face_landmarks, mp_holistic.FACEMESH_CONTOURS) # draw connection in results dictionar, face key that is face conncetion
    mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS)
    mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS)
    mp_drawing.draw_landmarks(image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS)
```

```python
def draw_styled_landmarks(image,results):
    # draw face connections
    mp_drawing.draw_landmarks(image, results.face_landmarks, mp_holistic.FACEMESH_CONTOURS,
                             mp_drawing.DrawingSpec(color=(80,110,10), thickness=1, circle_radius=1), 
                             mp_drawing.DrawingSpec(color=(80,256,121), thickness=1, circle_radius=1)
                             )
    # draw pose connections
    mp_drawing.draw_landmarks(
                             image, results.face_landmarks, mp_holistic.POSE_CONNECTIONS,
                             mp_drawing.DrawingSpec(color=(80,22,10), thickness=2, circle_radius=4),
                             mp_drawing.DrawingSpec(color=(80,44,121), thickness=2, circle_radius=2)
                             )
    # draw left hand connections
    mp_drawing.draw_landmarks(
                             image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS,
                             mp_drawing.DrawingSpec(color=(121,22,76), thickness=2, circle_radius=4),
                             mp_drawing.DrawingSpec(color=(121,44,250), thickness=2, circle_radius=2)
                             )
    # draw right hand connections
    mp_drawing.draw_landmarks(
                             image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS,
                             mp_drawing.DrawingSpec(color=(245,117,66), thickness=2, circle_radius=4),
                             mp_drawing.DrawingSpec(color=(245,66,230), thickness=2, circle_radius=2)
                             )
                             
```

```python
cap = cv.VideoCapture(0)
# set mediapipe 
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holisitic:
    while cap.isOpened():
        ret, frame = cap.read()
        # Make detection 
        image, results = mediapipe_detection(frame,holisitic)
#         print(results)
        # draw landmarks
        draw_styled_landmarks(image, results)
        cv.imshow('OpenCV feed',image)
        if cv.waitKey(10) & 0xFF==ord('q'):
            break
cap.release()
cv.destroyAllWindows()
```

```python
def extract_keypoints(results):
    pose = np.array([[res.x, res.y, res.z, res.visibility] for res in results.pose_landmarks.landmark]).flatten() if results.pose_landmarks else np.zeros(33*4)
    face = np.array([[res.x, res.y, res.z] for res in results.face_landmarks.landmark]).flatten() if results.face_landmarks else np.zeros(468*3)
    lh = np.array([[res.x, res.y, res.z] for res in results.left_hand_landmarks.landmark]).flatten() if results.left_hand_landmarks else np.zeros(21*3)
    rh = np.array([[res.x, res.y, res.z] for res in results.right_hand_landmarks.landmark]).flatten() if results.right_hand_landmarks else np.zeros(21*3)
    return np.concatenate([pose, face, lh, rh])
```

Setup folders for data collection

```python
# path for exported data, numpy arrays
DATA_PATH = os.path.join('MP_Data')

# Actions that we are trying to detect
actions = np.array(['hello', 'thanks', 'iloveyou'])

# thirty videos of data
no_sequences = 30

# videos are going to be in frames of length
sequence_length = 30
```

```python
# we are going to create 30 videos per action i.e, hello, thanks, ily
# then each of the video sequence are going to contain 30 frames of data
# each of the frame will contain 1662 landmarks
```

```python
for action in actions:
    for sequence in range(no_sequences):
        try:
            os.makedirs(os.path.join(DATA_PATH, action, str(sequence)))
        except:
            pass
```

Collect vidoes of training and testing

```python
cap = cv.VideoCapture(0)
# set mediapipe 
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holisitic:
    while cap.isOpened():
        # loop through actions
        for action in actions:
            # loop through sequence aka videos
            for sequence in range(no_sequences):
                # loop through video length aka sequence length
                for frame_num in range(sequence_length):
                    
                    ret, frame = cap.read()
                    # Make detection 
                    image, results = mediapipe_detection(frame,holisitic)
                    # draw landmarks
                    draw_styled_landmarks(image, results)
                    # this will move too fast, so we will need to apply brakes after every sequence
                    
                    # apply wait logic
                    if frame_num==0:
                        cv.putText(image,'STARTING_COLLECTION',(120,200),cv.FONT_HERSHEY_SIMPLEX,1,(0,255,0),4,cv.LINE_AA)
                        cv.putText(image,'Collecting frames for {} video number {}'.format(action,sequence),(15,12),cv.FONT_HERSHEY_SIMPLEX,0.5,(0,0,255),1,cv.LINE_AA)
                        cv.waitKey(2000)
                    else:
                        cv.putText(image,'Collecting frames for {} video number {}'.format(action,sequence),(15,12),cv.FONT_HERSHEY_SIMPLEX,0.5,(0,0,255),1,cv.LINE_AA)
                    
                    # New export keypoints
                    keypoints = extract_keypoints(results)
                    npy_path = os.path.join(DATA_PATH, action, str(sequence), str(frame_num))
                    np.save(npy_path, keypoints)
                    
                    # Show to screen
                    cv.imshow('OpenCV feed',image)
                    
                    if cv.waitKey(10) & 0xFF==ord('q'):
                        break
        
        
cap.release()
cv.destroyAllWindows()
```

Preprocess data and create labels and features

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
```

```python
label_map = {label:num for num,label in enumerate(actions)}
label_map
```

```python
sequences, labels = [], [] 
for action in actions:
    for sequence in range(no_sequences):
        window = []
        for frame_num in range(sequence_length):
            res = np.load(os.path.join(DATA_PATH, action, str(sequence), "{}.npy".format(frame_num)))
            window.append(res)
        sequences.append(window)
        labels.append(label_map[action])
```

```python
np.array(sequences).shape
```

```python
x = np.array(sequences)
```

```python
y = to_categorical(labels).astype(int)
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.05)
```

Build and train LSTM neural network

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.callbacks import TensorBoard
```

```python
log_dir = os.path.join('Logs')
tb_callback = TensorBoard(log_dir=log_dir)
```

```python
model = Sequential()
# when we are stacking LSTM layers, we need to return sequences = true as the next layer needs it
model.add(LSTM(64, return_sequences=True, activation='relu', input_shape=(30,1662)))
model.add(LSTM(128, return_sequences=True, activation='relu'))
model.add(LSTM(64, return_sequences=False, activation='relu')) # next step not lstm, so no need of returning sequences
model.add(Dense(64, activation='relu'))
model.add(Dense(128,activation='relu'))
model.add(Dense(actions.shape[0], activation='softmax'))
```

```python
model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['categorical_accuracy'])
```

```python
model.fit(x_train, y_train, epochs=2000, callbacks=[tb_callback])
```

Test in Real time

```python
# new detection variables
sequence = []  # for adding keypoints
sentence = []  # to display predicted action
predictions = []
threshold = 0.4

cap = cv.VideoCapture(0)

# Set mediapipe model
with mp_holistic.Holistic() as holistic:
    while cap.isOpened():
        # Real feed
        ret, frame = cap.read()

        if not ret:
            break

        # Make detections
        image, results = mediapipe_detection(frame, holistic)

        # Draw landmarks
        draw_styled_landmarks(image, results)

        # 2. Prediction Logic
        keypoints = extract_keypoints(results)
        sequence.append(keypoints)
        sequence = sequence[-30:]  # grab last 30 frames for prediction

        # X.shape is (30,1662) but the model is expecting the shape (num_sequences,30,1662), so use
        if len(sequence) == 30:
            res = model.predict(np.expand_dims(sequence, axis=0))[0]
            print(actions[np.argmax(res)])

            # 3. Vis logic
            if res[np.argmax(res)] > threshold:  # it should be more than detection threshold
                if len(sentence) > 0:
                    if actions[np.argmax(res)] != sentence[-1]:  # the prediction hasnt been added yet, so add prediction
                        sentence.append(actions[np.argmax(res)])
                else:
                    sentence.append(actions[np.argmax(res)])  # add first prediction

            if len(sentence) > 5:
                sentence = sentence[-5:]  # only last 5 predictions be displayed

            
            cv.rectangle(image, (0, 0), (640, 40), (245, 117, 16), -1)
            cv.putText(image, ' '.join(sentence), (3, 30), cv.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 255), 1, cv.LINE_AA)

        # Show to screen
        cv.imshow('OpenCV feed', image)

        # Break
        if cv.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv.destroyAllWindows()
```
